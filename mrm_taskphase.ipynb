```python
import numpy as np
import matplotlib.pyplot as plt
import csv
%matplotlib inline
# Function to load CSV data
def load_data(csv_file):
    # Initialize lists to store the dataset
    features = []
    prices = []
    
    # Open the csv file and read its content
    with open(csv_file, 'r') as file:
        csv_reader = csv.reader(file)
        next(csv_reader)  # Skip the header
        for row in csv_reader:
            features.append(float(row[0]))  # Assuming the feature is in the first column
            prices.append(float(row[1]))    # Assuming the price is in the second column
    
    return np.array(features), np.array(prices)

# Function to perform linear regression
def linear_regression(X, y, learning_rate=0.01, iterations=1000):
    m = 0
    b = 0
    n = len(y)
    
    for _ in range(iterations):
        # Prediction
        y_pred = m * X + b
        
        # Compute the cost (Mean Squared Error)
        cost = (1/n) * sum((y_pred - y) ** 2)
        
        # Gradient Descent
        dm = -(2/n) * sum(X * (y - y_pred))
        db = -(2/n) * sum(y - y_pred)
        
        # Update parameters
        m = m - learning_rate * dm
        b = b - learning_rate * db
        
    return m, b

# Function to plot the results
def plot_regression_line(X, y, m, b):
    plt.scatter(X, y, color='blue', label='Data Points')
    y_pred = m * X + b
    plt.plot(X, y_pred, color='red', label='Regression Line')
    plt.xlabel('Feature (e.g., Age, Mileage, etc.)')
    plt.ylabel('Price')
    plt.title('Linear Regression for Car Prices')
    plt.legend()
    plt.show()

# Main program
csv_file = 'CarPrice_Assignment.csv'  # Path to your CSV file
X, y = load_data(csv_file)

# Perform linear regression
learning_rate = 0.01
iterations = 1000
m, b = linear_regression(X, y, learning_rate, iterations)

# Display results
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# Plot the regression line
plot_regression_line(X, y, m, b)
```
